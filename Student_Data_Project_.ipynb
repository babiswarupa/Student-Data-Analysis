```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfs=pd.read_csv('student_data.csv')
dfs
```

```python
dfs.info()
```

```python
dfs['TestPrep'][dfs['TestPrep'].isnull()]
```

```python
dfs.rename(columns={'WklyStudyHours':'weekly Studyhours'},inplace=True)
dfs
```

```python
dfs['weekly Studyhours'].value_counts()
```

```python
dfs.info()
```

```python
dfs['Averagescore']=dfs[['MathScore','ReadingScore','WritingScore']].mean(axis=1)
dfs
```

```python
plt.figure(figsize=(40,27),facecolor = 'lightcyan')
plt.suptitle('Student Data Analysis',fontsize=90,fontstyle='italic')
plt.subplot(3,3,1)
gc1=dfs['Gender'].value_counts()
gc1
ex=[0.15,0.01]
plt.pie(gc1,labels=gc1.index,autopct='%0.0f%%',colors=['blue','green'],shadow='True',startangle=75,radius=1,textprops={'fontsize':15,'color':'black','fontstyle':'italic'},pctdistance=0.83,explode=ex)
plt.pie([1],radius=0.7,colors=['lightcyan'])
ax=plt.gca()
ax.set_title('Gender Distribution on Students',fontsize=35,color='Black')

plt.subplot(3,3,2)
g=dfs['ParentEduc'].value_counts()
plt.pie(g.values,labels=g.index,autopct='%0.1f%%',shadow='True',colors=['red','orange','green','purple','cyan','brown','white'],startangle=75,radius=1,textprops={'fontsize':15,'color':'black','fontstyle':'italic'})
ax=plt.gca()
ax.set_title('Parent Education levels',fontsize=35,color='Black')

plt.subplot(3,3,3)
d=sns.barplot(x='ParentEduc',y='Averagescore',data=dfs,hue='Gender',palette='rocket',errorbar=None,estimator='sum',width=0.5)
for i in d.containers:
  d.bar_label(i,fmt='%d')
plt.gca().set_facecolor('silver')
ax=plt.gca()
ax.set_title('Parent education Vs Stud Performance ',fontsize=35,color='black')

plt.subplot(3,3,4)
a = dfs.pivot_table(values =['MathScore','ReadingScore','WritingScore'],index=['weekly Studyhours','Gender'],aggfunc=('sum','mean'))
sns.heatmap(a,annot=True,cmap='rainbow',linewidths=1.55,linecolor='black',fmt='.2f')
ax=plt.gca()
ax.set_title('Weekly Studyhours Vs Stud Performance ',fontsize=35)

plt.subplot(3,3,5)
e=sns.barplot(x='TestPrep',y='Averagescore',data=dfs,hue='Gender',palette='light:purple',errorbar=None,estimator='sum',width=0.5)
for i in e.containers:
  e.bar_label(i,fmt='%d')
plt.gca().set_facecolor('skyblue')
ax=plt.gca()
ax.set_title('Test Preparation Vs Stud Performance ',fontsize=35,color='black')



plt.subplot(3,3,6)
a = dfs.pivot_table(values =['MathScore','ReadingScore','WritingScore'],index='Gender',aggfunc=('sum','mean'))
sns.heatmap(a,annot=True,cmap='Blues',linewidths=1.55,linecolor='black',fmt='.2f')
ax=plt.gca()
ax.set_title('Gender-wise Performance',fontsize=35)

plt.subplot(3,3,7)
g=sns.barplot(x='PracticeSport',y='Averagescore',data=dfs,hue='Gender',palette='dark:red',errorbar=None,estimator='sum',width=0.5)
for i in g.containers:
  g.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightblue')
ax=plt.gca()
ax.set_title('Sports Practice Vs Stud Performance',fontsize=35,color='black')

plt.subplot(3,3,8)
a = dfs.pivot_table(values =['MathScore','ReadingScore','WritingScore'],index=['EthnicGroup','Gender'],aggfunc=('sum','mean'))
sns.heatmap(a,annot=True,cmap='Greens',linewidths=1.55,linecolor='black',fmt='.2f')
ax=plt.gca()
ax.set_title('EthnicGroup-wise Performance',fontsize=35)

plt.subplot(3,3,9)
g=sns.barplot(x='LunchType',y='Averagescore',data=dfs,hue='Gender',palette='twilight',errorbar=None,estimator='sum',width=0.5)
for i in g.containers:
  g.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightgrey')
ax=plt.gca()
ax.set_title('LunchType Vs Stud Performance',fontsize=35,color='black')

plt.tight_layout()
plt.show()



```

**Conclusion:**

**1.Gender Distribution on Students:-**

The pie chart shows that the gender distribution among students is **equal —50% male and 50% female**. This indicates a **balanced representation** of both genders in the student group.

**2.Parent Education levels:-**

This indicates the **majority of students** come from some college education(23%).

**3.Parent education Vs Stud Performance :-**

Students whose parents have **some college education **achieved the **highest average scores** among all education levels.

**4.Weekly Studyhours Vs Stud Performance:-**

Students who study** more than 10 hours per week** generally achieve higher overall scores across all subjects, showing that increased study time improves academic performance.

**5.Test Preparation Vs Stud Performance:-**

Students who **completed** the test preparation scored noticeably **higher on average than those who did not**, showing that test preparation significantly improves student performance.

**6.Gender-wise Performance:-**

**Female students **consistently **scored higher than male students** across Math, Reading, and Writing, indicating better overall academic performance by females.

**7.Sports Practice Vs Stud Performance:-**

 **students whosometimes practice sports have the highest average
scores**, indicating that moderate sports participation is linked to better academic performance.

**8.EthnicGroup-wise Performance:-**

**Group E Female Students** have higher Reading and Writing Scores compared to males across ethnic groups.

**9.LunchType Vs Stud Performance:-**

Students with **standard lunch perform **better than those with** free or reduced lunch**, and **females score slightly higher** than males overall.
